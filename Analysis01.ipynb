```python
import pandas as pd
dataset=pd.read_csv("/content/file.csv")
df=pd.DataFrame(dataset)
df.head()
```

```python
def avg(ds,col):
  s=0
  c=0
  for i in ds[col]:
    try:
      s=s+i
      c=c+1
    except:
      continue
  return s/c
print("Successful")
```

```python
df['user_rating']=df['user_rating'].fillna(avg(df,"user_rating"))
df['user_rating']
```

```python
df['watch_date']=pd.to_datetime(df['watch_date'],dayfirst=False)
df['watch_date']
```

```python
def watch_hours(ds,col):
  hour=[]
  for i in ds[col]:
    try:
      i=int(i)
      h=i/60
      hour.append(h)
    except:
      x=None
      hour.append(x)
  return hour
watch_hours(df,"watch_time_minutes")
df['watch_time_hours']=watch_hours(df,"watch_time_minutes")
df['watch_time_hours']
```

```python
df.to_csv("user_activity_warehouse.csv",index=False)
```
